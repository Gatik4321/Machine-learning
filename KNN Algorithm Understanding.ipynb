```python
dfx=pd.read_csv("../Datasets/xdata.csv")
dfy=pd.read_csv("../Datasets/ydata.csv")
```

```python
# converting the csv files into the numpy array
X=dfx.values
Y=dfy.values

Y=X[:,1:]# We are here including all the columns in our data set except the first colums
Y=Y[:,1:]# we are here inclidng all tge colums in our data ste excepr the seocnd columsn 

print(X)

print(X.shape)
print(Y.shape)
```

```python
# Now we are plotting the two classes one the columns 1 of class 1
# one is column 2 of class 2
query_x=np.array([2,3]) # this point is used to test for the KNN
plt.scatter(X[:0],X[:,1],c=Y) # from this command we have divide the given data set into classes
plt.show()
```

```python
# we are commputing the ecludian distance for that point from all the given neighbours
def dist(x1,x2):
    return np.sqrt(sum(x1-x2)**2)
def knn(X,Y,querypoint,k=5):
    vals=[]
    m=X.shape[0]
    
    for i in range(m):
        d=dist(querypoint,X[i])
        vals.append(d,Y[i])
    vals=sorted(vals)
    vals=vals[:k]
    
    vals=np.array(vals)
    print(vals)
    new_vals=np.unique(vals[:,1],return_counts=True)
    print(new_vals)
    
    index=new_vals[1].argmax()
    pred=new_vals[0][index]
    return pred
```

```python
x=knn(X,Y,query_x)
print(x)
np.unique(vals,return_counts=True)
```

# Project of recognizing the misit data set using KNN

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('mnist_train.csv.csv')
print(df.shape)
```

```python
print(df.columns)
```

```python
df.head()
```

```python
data=df.values
```

```python
print(data.shape)
print(type(data))
```

```python
X=data[:,1:]
Y=data[:,0]
```

```python
print(X.shape)
print(Y.shape)
```

```python
# Now spliting the data into the training and the testing part
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,random_state=100)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#visulalize some samples
def drawImg(sample):
    img=sample.reshape((28,28))
    plt.imshow(img,cmap='gray')
    #plt.show()
drawImg(X_train[190])
```

```python
# can we apply the KNN to this data
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

# Now writing the logic for the kNN
def knn(X,Y,queryPoint,k=5):
    vals=[]
    m=X.shape[0]
    
    for i in range(m):
        d=dist(queryPoint,X[i])
        vals.append((d,Y[i]))
        
    vals=sorted(vals)
    vals=vals[:k]
    vals=np.array(vals)
    
    new_vals=np.unique(vals[:,1],return_counts=True)
    
    index=new_vals[1].argmax()
    pred=new_vals[0][index]
    
    return pred
```

```python
pred=knn(X_train,Y_train,X_test[0])
print(int(pred))
```
