```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression 
```

```python
X,y=make_regression(n_samples=1000,n_features=2,n_informative=1,noise=10,random_state=1)
```

```python
print(X.shape,y.shape)
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)

plt.subplot(1,2,2)
plt.scatter(X[:,1],y)
```

```python
# Now we learn how to plot 3d plot using matplot lib
%matplotlib notebook
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(10,7))
ax=plt.axes(projection='3d')

# creating plot
ax.scatter3D(X[:,0],X[:,1],y,color="green")
plt.title("3D plot scatter")
plt.show()
```

```python
# Now implementing the linear regression using sklearn
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([X[0],X[1]])
```

```python
model.score(X,y)
```
